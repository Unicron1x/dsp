```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
```

```python
import thinkstats2

random_numbers = np.random.random(1000)

pmf = thinkstats2.Pmf(random_numbers)
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='Random Number', ylabel='Pmf')
```

```python
cdf = thinkstats2.Cdf(random_numbers)
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel = 'Random Numbers', ylabel='CDF')
```
